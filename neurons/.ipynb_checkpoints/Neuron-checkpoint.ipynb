```python
def g(u,s):
    if u>s:
        return 1
    if u <= s:
        return -1
    else:
        return
def h(x_initial, y_initial, n, beta, s, excite = True):
    x=[]
    y=[]
    k=len(y_initial)
    for item in x_initial:
        x.append(item)
    for item in y_initial:
        y.append(item)
    for i in range(n):
        step=len(x)-1
        x.append(beta*x[step]-g(y[step-k],s))
        y.append(beta*y[step]+g(x[step-k],s))
    return [x,y]
```
