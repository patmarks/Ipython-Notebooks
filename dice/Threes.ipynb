⚂ The Game of Threes ⚂
===
<img src="dice1.png">

Threes is a game played with 5 dice, and two or more players. While the rules are simple, it's endlessly amusing. In a single round, each roller makes a play trying to obtain the lowest score. The end score is the sum of the dice the player holds, with 3's being counted as zero. A perfect score is thus (3,3,3,3,3), and the worst score is (6,6,6,6,6).

1. All dice are rolled.
2. Of those dice, the roller must keep one or more for scoring
3. With the remaining dice (if any), go back to step one.
4. Once all 5 dice are being held, they're added up with 3's being counted as zero.

An example play:
* Roll1: ⚅ ⚃ ⚀ ⚂ ⚃ ⚁
* keep: ⚂ ⚀
* score:⚂ ⚀ (1)


* Roll2:⚃ ⚅ ⚁
* keep: ⚁
* score:⚂ ⚀ ⚁ (3)


* Roll3:⚀ ⚁
* keep:⚀ ⚁
* score:⚂ ⚀ ⚁ ⚀ ⚁ (6)





The Brief Analysis
===
This notebook's purpose is to briefly look at how scoring 3 as 0 affects the gameplay.

First, a roll function is defined. This takes an array of scores as input, and outputs the possible scores with rolling one more die. The zilch parameter will score that number as a zero. Thus, roll([0], zilch = 3) will give an output of [1,2,0,4,5,6].roll([0], zilch = None) will give an output of [1,2,3,4,5,6], essentially getting rid of the titular threes rule.

Then, a function is defined to recursively apply the roll function to [0]. As an input, the recursive function takes the number of dice being rolled. It outputs all of the possible scores. The size of the output array is 6^n, where n is the number of dice.

Lastly, a function for making a histogram of the possible scores is defined.

```python
import matplotlib.pyplot as plt
%matplotlib inline

die = range(1,7)

#The roll functions take an array of scores as input
#and outputs an array of all the possible new scores
def roll(scores, zilch = 3):
    out=[]
    for score in scores:
        for roll in die:
            if roll == zilch:
                roll = 0
            out.append(score+roll)
    return out


#the recursive functions simply feed the roll functions
#a seed of [0], and then recursively apply the roll
#The output is an array 6^number_of_dice long
def roll_recursive(number_of_dice,zilch=3):
    out=[0]
    for n in range(1,number_of_dice+1):
        out = roll(out,zilch)
    return out




def histogram(set):
    histo  = []
    ways = 0
    possible_scores=range(max(set)+1)
    for n in possible_scores:
        total = 0
        for i in set:
            if i == n:
                total += 1
        histo.append(total)
    return histo

def cumulate(function):
    total = 0
    output=[]
    for i in function:
        total = total + i #function[i];
        #output.append(total);
        output=np.append(output,total)
    for n in range(len(output)):
        output[n]=output[n]/output[-1]
    return output
```

Here, we define the histograms for a rolls with 2 and 5 dice, with and without the threes rule.

```python
twodicenormal=roll_recursive(2,zilch=None)
twodicethrees=roll_recursive(2,zilch = 3)

histwothrees = histogram(twodicethrees)
histwonorm = histogram(twodicenormal)

possible_scores = range(6*2+1)

plt.rc('font', size = 18)

plt.figure(figsize=(10,7))
plt.clf()
plt.plot(possible_scores,histwothrees,label='Threes rule',lw=5)
plt.plot(possible_scores,histwonorm,label='Normal Dice',lw=5)
plt.xlabel('Score')
plt.ylabel('Frequency of Score')
plt.title('Two Dice')
plt.axis([0,12,0,7])
plt.legend(loc=2)
plt.show()

print str('\n\n\n\n\n')

fivedicenormal=roll_recursive(5,zilch=None)
fivedicethrees=roll_recursive(5,zilch = 3)

his5threes = histogram(fivedicethrees)
his5norm = histogram(fivedicenormal)

plt.figure(figsize=(10,7))

possible_scores = range(max(fivedicenormal)+1)
plt.clf()
plt.plot(possible_scores,his5threes,label='Threes rule',lw=5)
plt.plot(possible_scores,his5norm,label='Normal Dice',lw=5)
possible_scores = range(max(twodicenormal)+1)


plt.xlabel('Score')
plt.ylabel('Frequency of Score')
plt.title('Five Dice')
#plt.axis([0,12,0,7])
plt.legend(loc=2)
plt.show()
```
