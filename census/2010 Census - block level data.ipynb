2010 Block-level Census Data
===
<img src="examples/boswash small.png">

The above image is a rendering of population density for every block from Boston to Washington, DC. This is known as the Boswash corridor. The level of detail is such one can zoom in enough to see individual neighborhoods and parks in the eastern seaboard. 

The motivation for this project was that, despite a variety of census tools online, there were none for visualization at the block level.  

Data needed in addition to this notebook:
This  notebook requires downloading each file for states of interest: 
1. The [sf1](http://www2.census.gov/census_2010/04-Summary_File_1/) file.
2. The [shapefile](http://www2.census.gov/geo/tiger/TIGER2010/TABBLOCK/2010/) tl_2010_xx_tabblock10.zip, where xx is the two digit [FIPS code](http://www.census.gov/geo/reference/ansi_statetables.html) for the state. 
3. The data field descriptors file, which is in this repository. 

The contents of each zip file must be organized as follows. In the notebook folder is a 'census data' folder. In the 'census data' folder, there is a '2010census' folder. Each state has its own folder in 'census data/2010census/2010census' root folder. The state folder is the lower case two letter abbreviation. In each state folder, there's a folder for the sf1 files called 'sf1', and then there's the folder with the shapefile. with a more messy name. For the block-level data, the folder is 'tl_2010_XX_tabblock10' where XX is the census two number abbreviation of the state.

# Reading input files and organizing data

```python
import csv
import numpy as np

import shapefile
import timeit
import heapq

import matplotlib.pyplot as plt
from matplotlib.collections import PolyCollection
import matplotlib.colors as colors
import matplotlib.cm as cm
```

### Classes

The StateHeaderAndShape class contains all of the shapefile data as StateHeaderAndShape.xy and StateheaderAndShape.xy_geoid. These are from the TIGER shapefile alone. However, in order to address these shapes with the sf1 file, you need to know the logical record number of every shape, not just the geoid. For that information, the geographical header in the sf1 file is needed. The header info in in StateHeaderAndShape.header. A dictionary  locrecdict in the header object converts the logical record number to the geoid.

When creating a StateHeaderAndShape object, the only two arguments are two strings for locating the files.

Each object takes a significant chunk of memory, since it's loading every single block of the state into the memory. On my computeer with a AMD FX-6300 CPU and 16gb of RAM, Rhode Island takes about 3 seconds to load, while New York takes a minute.

```python
census2010folder='census data/2010census/'

class StateHeaderAndShape:
    """
    This class creates an object for each state, which contains all geographic information
    
    geo_header_file is the raw data from the geo header within the sf1 package

    header is selected data from the header file.
    Most importantly, it has the following dictionaries:
        geo_dict
        logrecno_dict
        
    
    string is the two character string for the state, which is used in file paths
    
    xy contains every polygon within the shape file
    """
    def __init__(self,state,state_num):
        
        class ParsedGeoHeader:
            """
            This object contains select data from the geo header file. The only argument
            is the path to the geo header. Each property is ordered by the LOGRECNO
            """
            def __init__(self, geo_file):
                def geo_column(starting_position, field_size, return_numbers = True):
                    """This function returns a list for the given column in the geo header file. 
                    The starting position is the starting record number, as in the data dictionary 
                    (which starts at 1)"""
                    col_start = starting_position-1 
                    # 1 is subtracted to convert from the data dictionary start to a list index
                    out=[]
                    col_end=col_start+field_size #field size from the data dictionary

                    if return_numbers == True:
                        for row in geo_file:
                            try:
                                out.append(float(str(row[0][col_start:col_end])))
                            except ValueError:
                                out.append(str(row[0][col_start:col_end]))
                    else:
                        out=[]
                        for row in geo_file:
                            out.append(str(row[0][col_start:col_end]))
                    return out
                ## Selected data points
                self.sumlev = geo_column(9,3, return_numbers=False)#summary level
                self.logrecno = geo_column(19,7, return_numbers = True)
                self.state_county = geo_column(28,5,return_numbers = False) # state and county
                self.census_tract = geo_column(55,6,return_numbers = False) # census tract
                self.block_group = geo_column(61,1,return_numbers = False) # block group
                self.block = geo_column(62,4,return_numbers = False) # block
                #The geoid is constructed from the state, county, census tract and block
                self.geoid = []
                for i in range(len(geo_file)):
                    self.geoid.append(self.state_county[i]+self.census_tract[i]+self.block[i])


                
                #position
                self.lat = geo_column(337,11)
                self.lon = geo_column(348,12)
                #AREA CHARACTERISTICS
                self.area_land = geo_column(199,14, return_numbers = True) #AREALAND
                self.area_water = geo_column(213,14, return_numbers =True ) #AREAWATR
                self.pop100 = geo_column(319,9, return_numbers = True) #POP100
                self.housing_units = geo_column(328,9, return_numbers =True ) #HU100

                self.length = len(self.logrecno)
                self.rang = range(len(self.logrecno))
                
                
                #make a dictionary for logrecno -> geoid
                self.logrecdict=dict()
                self.pop100dict=dict()
                for logrecno, geoid,pop100 in zip(self.logrecno,self.geoid,self.pop100): 
                    self.logrecdict[geoid]=logrecno
                    self.pop100dict[geoid]=pop100
                    
                    
                    
                
                self.geodict=dict()
                for logrecno, geoid in zip(self.logrecno,self.geoid):
                    if geoid[0]!=' ':
                        self.geodict[logrecno]=geoid
        
        #
        #     
        #       
        #READ SHAPEFILE
        start_time = timeit.default_timer()

        self.dat=shapefile.Reader(census2010folder+state+
            '/tl_2010_'+state_num+'_tabblock10/tl_2010_'+state_num+'_tabblock10')
        self.info=dict()
        for i in range(len(self.dat.fields)-1):
            self.info[self.dat.fields[i+1][0]]=i
        
        
        #copy the shape data into xy, xy_info and xy_geoid
        #xy_geoid info is contained in xy_info, but the lists are sorted
        #using xy_geoid, so that a binary search can be used when making a map
        self.xy=[]
        self.xy_info=[]
        self.xy_geoid=[]
        self.area = []
        for shape, data in zip(self.dat.shapes(),self.dat.records()):
            if data[self.info['ALAND10']]>0: # eliminates places with no land.
                for i in range(len(shape.parts)): 
                    #shape.parts contains the starting points for each polygon within the points
                    #this for loop creates every polygon to be mapped
                    self.xy_geoid.append(data[self.info['GEOID10']].ljust(15))
                    self.area.append(int(data[self.info['ALAND10']]))
                    if shape.parts[i] == shape.parts[-1]:
                        self.xy.append(shape.points[shape.parts[i]:])
                    else:
                        self.xy.append(shape.points[shape.parts[i]:shape.parts[i+1]])
        self.dat=None
                
        #END shapefile reading
        #
        #
        #
                
        self.string = state
        
        print "read shapefile in "+str(
            timeit.default_timer() - start_time)[0:3]+'s'

        
        #from here, the geo header file is read
        start_time = timeit.default_timer()
        self.geo_header_file=[]
        #copy contents of the header into self.geo_header_file
        with open(str(census2010folder+state+'/sf1/'+state+'geo2010.sf1'),'rb') as f:
            reader = csv.reader(f,delimiter='*', quoting=csv.QUOTE_MINIMAL)
            for row in reader:
                self.geo_header_file.append(row)
        # parse the header with the ParsedGeoHeader class
        self.header = ParsedGeoHeader(self.geo_header_file)
        self.geo_header_file = None
        
        print "read header file in "+str(
            timeit.default_timer() - start_time)[0:3]+'s'
```

```python
class Data_Dict():
    """
    The Data_Dict codedict dictionary takes census data codes and returns the
    file number and column position.
    
    It's a class rather than a simple dictionary for the possibility of making a 
    hierarchical structure for browsing what's in the DATA_FIELD_DESCRIPTORS2.txt
    file
    """
    def __init__(self, descriptor_file=census2010folder+'DATA_FIELD_DESCRIPTORS2.txt'):
            
        self.description=[] #copy the file to self.description
        with open(descriptor_file,'rb') as f:
            reader = csv.reader(f,delimiter=',', quoting=csv.QUOTE_MINIMAL)
            for row in reader:
                self.description.append(row)
                
        self.codedict = dict()
        self.heading_number=0
        self.heading=list()
        segment = 0 #this is equivalent to the segment of the sf1 file 
        record = 5 #the record in the sf1 file 
        for row,i in zip(self.description[1:],range(len(self.description))[1:]):
            if int(row[1]) > segment:
                segment += 1
                record = 5
            if (row[4]==''):
                if self.description[i-1][4]=='':
                    self.heading[self.heading_number-1].append(row[3])
                    
                else:
                    self.heading.append([row[3]])
                    self.heading_number += 1
            if row[4] != '': #row[1] is the segment, row[4] is the field code
                self.codedict[row[4]] = (row[1],record)
                self.heading[self.heading_number-1].append([row[3],row[4]])
                record += 1

codes=Data_Dict()

#Reads the given sf1 file, and outputs specified column to a numpy array
def sf1_column(state,field_code, return_numbers=True):
    """
    Takes as an input the state and sf1 field code.
    
    Returns a dictionary in which the keys are logical record numbers
    and the values are the values for the sf1 field. 
    """
    start_time = timeit.default_timer()
    sf1=[]
    segment,column=codes.codedict[field_code]
    #read document into sf1
    with open(census2010folder+state+'/sf1/'+state+'000'+\
                  segment.zfill(2)+'2010.sf1','rb') as f:
        reader = csv.reader(f,delimiter=',', quoting=csv.QUOTE_MINIMAL)
        for row in reader:
            sf1.append(row)

    
    
    data=[]
    logrecno=[]
    for line in sf1:
        data.append(line[column])
        logrecno.append(line[4])
    print "read sf1 file in "+str(
        timeit.default_timer() - start_time)[0:3]+'s'
    
    out=dict()
    for datum,num in zip(data,logrecno):
        if return_numbers==True:
            out[int(num)]=float(datum)
        else:
            out[int(num)]=datum
    return out
```

```python
def plot_map(place,datacode='P0010001',axes=[],datadict=dict(),size=10,bg=(0,0,0),
             cmap=cm.ocean, vmaxpercentile=0.1,vminpercentile=0.9,alpha=1, density=False,
             percentage=False, inline=False):
    """
    takes a list of StateHeaderAndShape objects, and a sf1 data code, and plots a map.
    By default, plots a 
    
    Yes, this function is overengineered. I thought there would be a lot to do with it,
    but there isn't much info at the block-level. Aspect ratio needs to be played with, 
    so inline matplotlib isn't the best choice.
    """
    start_time = timeit.default_timer()

    data=[]
    shapes=[]
    for state in place:
        statedata=[]
        if datadict!=dict(): 
            for geoid in state.xy_geoid:
                statedata.append(datadict[state.header.logrecdict[geoid]])
        else:
            rawdata=sf1_column(state.string,datacode)
            for geoid in state.xy_geoid:
                statedata.append(rawdata[state.header.logrecdict[geoid]])

        if density==True:
            new_data=[]
            for item, area in zip(statedata, state.area):
                new_data.append(float(item)/float(area))
            statedata=new_data
        if percentage==True:
            new_data=[]
            excepts=0
            for datum, geoid in zip(statedata, state.xy_geoid):
                try:
                    new_data.append(float(datum)/float(state.header.pop100dict[geoid]))
                except:
                    excepts+=1
                    new_data.append(0.)
            statedata=new_data
            print excepts
        shapes=shapes+state.xy
        data=data+statedata
        
    print "parsed sf1 file in "+str(
        timeit.default_timer() - start_time)[0:3]+'s'
        
        
    #if axes!=[]:
     #   size=(20,20)
        #ratio=(axes[1]-axes[0])/(axes[3]-axes[2])
        #size=(size[0]/ratio,size[0]*ratio)
    if inline==False:
        %matplotlib
        fig=plt.figure()
    if inline==True:
        %matplotlib inline
        fig=plt.figure(figsize=size)
    if inline=='save':
        fig=plt.figure(figsize=size)        

    # this is the default maximum value for coloring the data. heapq.nlargest makes a list 
    # of values in the percentile given. The lowest of those is then chosen as the max color value.

    if type(vmaxpercentile) is float:
        vmax = heapq.nlargest(int(vmaxpercentile*len(data)),data)[int(vmaxpercentile*len(data))-1] 
    if vmaxpercentile is 'max':
        vmax = max(data)
    if percentage==True:
        vmax=1
    print str(vmax*2.59*10**6)+'people per sq mile'
    if vminpercentile is 'min':
        vmin = min(data)
    if type(vminpercentile) is float:
        vmin = heapq.nsmallest(int(vminpercentile*len(data)),data)[0]
    ax = plt.subplot(1,1,1, axisbg = bg)
    norm = colors.Normalize(vmin,vmax,clip=True)
    poly = PolyCollection(shapes,color=cmap(norm(data)),alpha=alpha)
    plt.gca().add_collection(poly)
    if axes==[]:
        plt.autoscale()
    else:
        plt.axis(axes)
    if inline=='save':
        plt.savefig('asd')
    #mpld3.display(fig)
  #  print "Made map in " +str(
  #      timeit.default_timer() - start_time)[0:3]+"s"
```

Starting Small with Rhode Island
===

Here, we make the RI StateHeaderAndShape object.

```python
RI = StateHeaderAndShape('ri','44')
```

Now, the map is created with the plot_map function. For this block-level data, it's most interesting when zooming in all the way, so the tk matplotlib interface is better for producing an image. However, it's still possible to get good results with the inline matplotlib.

```python
#Density = True divides the data by the area.
#vmaxpercentile is the percentile of data that is colored as pure white. 
#So at 0.01, only 1 percent of the shapes are colored pure white.

#P0010001 is the data code for total population
plot_map([RI],datacode='P0010001',size=(15,20),density=True,bg=(0,0,0),
         cmap=cm.gray,vmaxpercentile=.01,
         vminpercentile='min', inline=True)
```

The axes keyword can be used to zoom into the picture. Here's a detail of the capital city, Providence. You can see the large swaths of commercial and industrial districts, and highways that carve the city neighborhoods up.

###### Zooming in

```python
plot_map([RI],datacode='P0010001',axes=[-71.5,-71.35,41.75,41.9],size=(15,15),density=True,bg=(0,0,1),
         cmap=cm.gray,vmaxpercentile=.01, inline=True)
```

###### Demographics

We can also look at percentages with this data to look at things like racial demographics of the area. P0040003 is the datacode for all "Hispanic or Latino" people. density = False, and percentage=True.

```python
plot_map([RI],datacode='P0040003',axes=[-71.5,-71.35,41.75,41.9],size=(15,15),percentage=True,density=False,bg=(0,0,1),
         cmap=cm.gray, vminpercentile='min', inline=True)
```
